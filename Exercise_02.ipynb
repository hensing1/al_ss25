# Exercise 2 - 1D Cellular Automaton

```python
import random

def is_valid_rule(rule: str, r: int, alphabet: [chr]) -> bool:
    if len(rule) != 2 ** (2 * r + 1):
        return False
    for letter in rule:
        if letter not in alphabet:
            return False
    return True


def str2int(input: str) -> int:
    if input.startswith("0b"):
        return int(input[2:], 2)
    if input.startswith("0x"):
        return int(input[2:], 16)
    return int(input)


def get_rule_index(neighborhood: str, alphabet: [chr]) -> int:
    index = 0
    l = len(neighborhood)
    for i in range(l):
        index += ((2 ** i) if neighborhood[l - i - 1] == alphabet[0] else 0)
    return index


def step(old_cells: str, alphabet: [chr], rule: str, r: int) -> str:
    new_cells = list(alphabet[0] * 84)
    for i in range(2, 82):
        neighborhood = old_cells[i-r : i+r+1]
        new_cells[i] = rule[get_rule_index(neighborhood, alphabet)]
    return "".join(new_cells)

def simulation(init: chr, alphabet: [chr], rule: str, steps: int = 42):
    if (init == "r"):
        cells = [0, 0] + [random.randint(0, 1) for _ in range(80)] + [0, 0]
    else:
        cells = [0] * 84
        cells[42] = 1

    cells = "".join([alphabet[i] for i in cells])

    print(cells)
    for _ in range(42):
        cells = step(cells, alphabet, rule, r)
        print(cells)

alphabet = [" ", "#"]
```

```python
r = 0
while (r != 1 and r != 2):
    r = int(input("Enter neighborhood (1 or 2): "))

rule = ""
while not is_valid_rule(rule, r, alphabet):
    rule_wolfram = str2int(input("Enter wolfram number: "))
    rule = bin(rule_wolfram)[2:].zfill(2 ** (2 * r + 1)) \
        .replace("0", alphabet[0]) \
        .replace("1", alphabet[1])

init = ""
while (init != "s" and init != "r"):
    init = input("Enter starting condition (s or r): ").lower()
    
simulation(init, alphabet, rule)
```
